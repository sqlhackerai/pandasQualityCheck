## read the csv file

```python
import pandas as pd
df = pd.read_csv('persons.csv')
```

```python
df
```

## Write functions to check data against these criteria.

```python
#Write functions to check data against these criteria. For instance

def validate_year_of_birth(year):
    return 1970 <= year <= 2021

def validate_blood_type(btype):
    return btype in ['A+', 'A-', 'B+', 'B-', 'O+', 'O-', 'AB+', 'AB-']

import re
def validate_social_security(ssn):
    return bool(re.match(r'^\d{3}-\d{2}-\d{4}$', ssn))
```

## adding quality columns

```python
#Apply Validations to the DataFrame: Use these functions to check the accuracy of data in each column:
df['ValidYearOfBirth'] = df['year_of_birth'].apply(validate_year_of_birth)
df['ValidBloodType'] = df['blood_type'].apply(validate_blood_type)
df['ValidSSN'] = df['social_security'].apply(validate_social_security)
```

## print dataframe

```python
df
```

## Report and Adjust in percent format

```python
invalid_years = df[df['ValidYearOfBirth']]
invalid_blood_types = df[df['ValidBloodType']]
invalid_ssns = df[df['ValidSSN']]

print(f"Invalid Year Count: {len(invalid_years)}")
print(f"Invalid Blood Types Count: {len(invalid_blood_types)}")
print(f"Invalid SSNs Count: {len(invalid_ssns)}")
```

## Report and Adjust in percent format

```python
total_records = len(df)

invalid_year_percentage = (len(invalid_years) / total_records) * 100
invalid_blood_type_percentage = (len(invalid_blood_types) / total_records) * 100
invalid_ssn_percentage = (len(invalid_ssns) / total_records) * 100

print(f"Invalid Year Count: {len(invalid_years)}, Percentage: {invalid_year_percentage:.2f}%")
print(f"Invalid Blood Types Count: {len(invalid_blood_types)}, Percentage: {invalid_blood_type_percentage:.2f}%")
print(f"Invalid SSNs Count: {len(invalid_ssns)}, Percentage: {invalid_ssn_percentage:.2f}%")
```
